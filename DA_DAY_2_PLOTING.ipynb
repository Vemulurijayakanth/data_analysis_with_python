```python
year=["19-20","20-21","21-22","22-23"]
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("Revenues of Company A",A)
print("Revenues of Company B",B)
c=np.diff(A)
d=np.diff(B)
print("profit of a",c)
print("profit of a",d)

plt.subplot(1,2,1)

plt.bar(year,c,label="company A",color="r")
plt.legend()
plt.title("profits of Company A & B")

plt.subplot(1,2,2)
plt.bar(year,d,label="company B",color="g")
plt.legend()
```

```python
#explode
a= np.array([25,60,5,10])
labe=["AIML","PYTHON","PANDAS","NUMPY"]
explo =[0.2,0,0,0]

plt.pie(a, labels=labe ,explode=explo, startangle=180 )
plt.legend(loc="best")

plt.show()
```
